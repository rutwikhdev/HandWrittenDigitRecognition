## Handwritten digit recognition on MNIST dataset using a Deep Neural Network

### First let's import all the necessary modules

```python
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
from DeepNN import *
from helper_utils import print_image
import pickle

%matplotlib inline
```

### Then let's load the dataset in (X, Y)train and (X, Y)test

```python
mnist = MNIST('../dataset/MNIST/')
x_train, y_train = mnist.load_training() #60000 samples
x_test, y_test = mnist.load_testing()    #10000 samples

x_train = np.reshape(np.asarray(x_train).astype(np.float32), (60000, 784))  # dim: (60000, 784) i.e. 60000 images of 28*28
y_train = np.reshape(np.asarray(y_train).astype(np.float32), (60000, 1))  # dim: (60000, 1) i.e. one vector with the correct value of digit (0-9)
x_test = np.reshape(np.asarray(x_test).astype(np.float32), (10000, 784))    # same as x_train with 10000 samples
y_test = np.reshape(np.asarray(y_test).astype(np.float32), (10000, 1))    # same as y_train with 10000 correct values
```

```python
index = 0              # change this index to see different image, between(0-59999)
print_image(x_train[index], y_train[index][0])
```

### Training Model

```python
network = np.array([784, 16, 16, 10])       # Linear-->Relu  Linear-->Relu  Linear-->Softmax
parameters = initi_params(network)
```

```python
one_hot_y_train = one_hot_matrix(y_train.squeeze())
one_hot_y_test = one_hot_matrix(y_test.squeeze())
```

```python
x_train = x_train / 255 #normalizing the data
t_params = L_layer_model(x_train.reshape(60000,784).T , one_hot_y_train, [784, 16, 16, 10], num_iterations=3000, learning_rate=0.15)
```

```python
trained_params = t_params
```

### Predicting on trained parameters

```python
x_test = x_test / 255

predictions_ts, caches_ts = L_model_forward(x_test.T, trained_params)
print("Test set predictions")
correct_test, incorrect_test = predict(one_hot_y_test.T, predictions_ts.T, 10000)
print("Test accuracy: ",correct_test*100 /10000, "%")

print("\n")

predictions_tr, caches_tr = L_model_forward(x_train.T, trained_params)
print("Train set predictions")
correct_train, incorrect_train = predict(one_hot_y_train.T, predictions_tr.T, 60000)
print("Test accuracy: ",correct_train*100 /60000, "%")
```
